### 1. What types are involved in the following expressions? What data type will each expression evaluate to?

`8 * 2.5`

```python
# integer * float operation will result in float
print(f'8 is of type:{type(8)}')
print(f'2.5 is of type:{type(2.5)}')
print(f'8 * 2.5 = {8 * 2.5} is of type:{type(8 * 2.5)}')
```

`9 / 2`

```python
# integer / integer operation will result in float
print(f'9 is of type:{type(9)}')
print(f'2 is of type:{type(2)}')
print(f'9 / 2 = {9 / 2} is of type:{type(9 / 2)}')
```

`9 // -2`

```python
# integer // integer operation will result in integer 
print(f'9 is of type:{type(9)}')
print(f'2 is of type:{type(2)}')
print(f'9 // (-2) = {9 // (-2)} is of type:{type(9 // (-2))}')
```

`1.5 * 2 >= 7 - 3`

```python
# float * integer operation will result in float
print(f'1.5 is of type:{type(1.5)}')
print(f'2 is of type:{type(2)}')
print(f'1.5 * 2 = {1.5 * 2} is of type:{type(1.5 * 2)}')

# integer - integer operation will result in integer
print(f'7 is of type:{type(7)}')
print(f'3 is of type:{type(3)}')
print(f'7 - 3 = {7 - 3} is of type:{type(7 - 3)}')

# float >= integer operation will result in boolean
print(f'1.5 * 2 >= 7 - 3 is {1.5 * 2 >= 7 - 3} of type:{type(1.5 * 2 >= 7 - 3)}')
```

### 2. In which order will the expressions be evaluated.

`6 * 3 + 7 * 4`


<details>
  <summary>Answer</summary>

  `*` > `*` > `+` 
</details>


`5 - 2 * 3 ** 4`

<details>
  <summary>Answer</summary>

  `**` > `*` > `-`
</details>

`(5 - 2) * 3 ** 4`

<details>
  <summary>Answer</summary>

  `(-)` > `**` > `*` 
</details>

`5 + 2 >= 3 * 4`

<details>
  <summary>Answer</summary>

  `*` > `+` > `>=`
</details>

```python
value = 6 * 3 + 7 * 4
print(f'6 * 3 + 7 * 4 ---> 18 + 28 = {value}') 

value = 5 - 2 * 3 ** 4
print(f'5 - 2 * 3 ** 4 ---> 5 - 2 * 81 ---> 5 - 162 = {value}') 

value = (5 - 2) * (3 ** 4)
print(f'(5 - 2) * (3 ** 4) ---> 3 * 81 = {value}')

value = 5 + 2 >= 3 * 4
print(f'5 + 2 >= 3 * 4 ---> 7 >= 12 {value}')
```
